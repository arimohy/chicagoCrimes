```python
#Ejercicio 3
def indiceCriminalidadArea(numeroDelitos,poblacion):
    if(poblacion>0):
        return numeroDelitos/poblacion*100
    else:
        print("No se puede hallar indice ya que la poblacion es igual a 0")

areas=[
    {
        'area':"Area 1",
        'numeroDelitos':289,
        'poblacion':54991
    },
    {
        'area':"Area 2",
        'numeroDelitos':228,
        'poblacion':71942
    },
    {
        'area':"Area 3",
        'numeroDelitos':319,
        'poblacion':56362
    },
    {
        'area':"Area 4",
        'numeroDelitos':141,
        'poblacion':39493
    },
    {
        'area':"Area 5",
        'numeroDelitos':5,
        'poblacion':0
    }
]
indices=[]
for a in areas:
    indice=indiceCriminalidadArea(a.get('numeroDelitos'),a.get('poblacion'))
    if(indice!=None):
        indices.append(indice)
    print(f"Indice del {a.get('area')} es {indice}")

print(f"Valor maximo {max(indices)}")
print(f"Valor minimo {min(indices)}")
print(f"Valor medio {sum(indices)/len(indices)}")
```
